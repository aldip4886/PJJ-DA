## **Cek Data** ##

### 1. Load data ke pandas

```python
# import pandas library
import pandas as pd

# declare link to dataset
link_data = "https://raw.githubusercontent.com/aldip4886/pjj_data_analytics/main/houseprice_train.csv"

# load data to pandas dataframe
df_data = pd.read_csv(link_data)
df_data.head()
#df_data
```

### 2. Cek apakah ada data yang memiliki data kosong/NaN, hitung berapa % data kosong


```python
# count missing values in each columns
missing_count = df_data.isnull().sum()

# calculate missing percentage
percent_missing = df_data.isnull().sum() * 100 / len(df_data)


df_missing_value = pd.DataFrame({'missing_count':missing_count,'percent_missing': percent_missing})
df_missing_value[(df_missing_value['missing_count']>0)]

```

### 3. Identifikasi kolom yang bersifat numerik dan non numerik

```python
#data types for each column
df_data.dtypes
```

```python
# get numeric columns
numeric = df_data._get_numeric_data().columns
print(numeric)
num_cols = df_data[numeric]

#count of numeric columns
print(str(num_cols.shape[1]) + " columns")
```

```python
# get non-numeric columns
non_numeric = df_data.drop(numeric, axis = 1).columns
print(non_numeric)
non_numeric_cols = df_data[non_numeric]

#count non-numeric columns
print(str(non_numeric_cols.shape[1]) + " columns")
```

## **Univariate** ##

### Cek Categorical Variables ###


#### 1. Buat barplot pada **MSZoning, Street, dan LotShape**


```python
# import matplot library
from matplotlib import pyplot as plt
```

```python
# summarize of MSZoning
df_data['MSZoning'].value_counts()
```

```python
# summarize of Street
df_data['Street'].value_counts()
```

```python
# summarize of LotShape

df_data['LotShape'].value_counts()
```

```python
#f,ax = plt.subplot(1,)
f,ax=plt.subplots(1,3,figsize=(15,4))

# subplot 1 = MSZoning
df_data['MSZoning'].value_counts().sort_index().plot.bar(color=['blue','green', 'red','yellow','magenta'],ax=ax[0])
ax[0].set_title('Zoning classification')
ax[0].set_ylabel('Count')


# subplot 2 = Street
df_data['Street'].value_counts().sort_index().plot.bar(color=['blue','green', 'red','yellow','magenta'],ax=ax[1])
ax[1].set_title('Type of road access to property')
ax[1].set_ylabel('Count')

# subplot 3 = LotShape
df_data['LotShape'].value_counts().sort_index().plot.bar(color=['blue','green', 'red','yellow','magenta'],ax=ax[2])
ax[2].set_title('General shape of property')
ax[2].set_ylabel('Count')

# show
plt.show()
```
