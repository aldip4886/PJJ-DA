```python
# jika ditampilkan maka akan keluar nilai x=10 dan nilai y=5

x = 5
y = 10

# ketik kode disini

y_temp = x
x_temp = y

y = y_temp
x = x_temp


print('nilai x :{}, dan nilai y:{}'.format(x,y))
```

```python
def swap (x,y):
    x = 5
    y = 10
    z = x
    x = y
    y = z

swap(5,10)
print('nilai x :{}, dan nilai y:{}'.format(x,y))
```
